```python
percentile25 = df1['fixed acidity'].quantile(0.25)
percentile75 = df1['fixed acidity'].quantile(0.75)
IQR = percentile75 -percentile25
upper_limit = percentile75 + 1.5 * IQR
lower_limit = percentile25 - 1.5 * IQR
```

```python
df2 = df1[(df1['fixed acidity'] <=upper_limit) & (df1['fixed acidity'] >=lower_limit)]
```

```python
df2['fixed acidity'].plot(kind ='box')
```
