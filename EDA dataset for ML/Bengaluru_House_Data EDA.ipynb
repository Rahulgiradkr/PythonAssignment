# Outlier removal

```python
B.boxplot(column=["price"])
```

```python
import numpy as np

def bhk_outlier_remove(B):
    exclude_indices = np.array([])
    for location, location_B in B.groupby("location"):
        bhk_stats = {}
        for bhk, bhk_B in location_B.groupby("BHK"):
            bhk_stats[bhk] = {
                "mean": np.mean(bhk_B.price_per_sqfeet),
                "std": np.std(bhk_B.price_per_sqfeet),
                "count": bhk_B.shape[0]
            }
            
        for bhk, bhk_B in location_B.groupby("BHK"):
            stats = bhk_stats.get(bhk)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_B[bhk_B["price_per_sqfeet"] < stats["mean"]].index.values)
                
    return B.drop(exclude_indices, axis="index")
```

```python
B = bhk_outlier_remove(B)
```
