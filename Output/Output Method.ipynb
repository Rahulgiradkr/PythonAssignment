```python
import logging

class HelloWorldPrinter:
    def __init__(self, message="Hello, World!"):
        self.message = message

    def print_hello_world(self):
        try:
            print(self.message)
        except Exception as e:
            logging.error(f"Error printing message: {e}")

if __name__ == "__main__":
    # Configure logging
    logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

    # Create an instance of the HelloWorldPrinter class
    hello_printer = HelloWorldPrinter()

    # Call the method to print "Hello, World!" and handle any exceptions using try-except
    hello_printer.print_hello_world()
```

```python
import logging

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class MultiplicationTablePrinter:
    def __init__(self):
        pass

    def print_multiplication_table(self, number):
        try:
            number = float(number)  # Convert input to float in case it's a decimal
            for i in range(1, 11):
                result = number * i
                print(f"{number} x {i} = {result}")
        except ValueError as ve:
            logging.error(f"Error: {ve}. Please enter a valid number.")
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    try:
        user_input = input("Enter a number: ")
        table_printer = MultiplicationTablePrinter()
        table_printer.print_multiplication_table(user_input)
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
```

```python
import logging

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class CountdownTimer:
    def __init__(self, start_number):
        self.start_number = start_number

    def start_countdown(self):
        try:
            for i in range(self.start_number, 0, -1):
                print(i)
            print("Blast-off!")
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")

    try:
        user_input = int(input("Enter the starting number for the countdown: "))
        countdown_timer = CountdownTimer(user_input)
        countdown_timer.start_countdown()
    except ValueError:
        logging.error("Error: Please enter a valid integer.")
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")

```

```python
import logging

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class StarPyramidGenerator:
    def __init__(self, num_rows):
        self.num_rows = num_rows

    def generate_pyramid(self):
        try:
            for i in range(1, self.num_rows + 1):
                spaces = " " * (self.num_rows - i)
                stars = "*" * (2 * i - 1)
                print(f"{spaces}{stars}")
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    try:
        user_input = int(input("Enter the number of rows for the pyramid: "))
        pyramid_generator = StarPyramidGenerator(user_input)
        pyramid_generator.generate_pyramid()
    except ValueError:
        logging.error("Error: Please enter a valid integer.")
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
```

```python
import logging

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class NumberPatternGenerator:
    def __init__(self, num):
        self.num = num

    def generate_pattern(self):
        try:
            for i in range(1, self.num + 1):
                pattern = ''.join(str(j) for j in range(1, i + 1))
                print(pattern, end=' ')
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    try:
        user_input = int(input("Enter a number 'n': "))
        pattern_generator = NumberPatternGenerator(user_input)
        pattern_generator.generate_pattern()
    except ValueError:
        logging.error("Error: Please enter a valid integer.")
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
```

```python
import logging

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class NumberSumCalculator:
    def __init__(self, num):
        self.num = num

    def calculate_sum(self):
        try:
            total_sum = sum(range(1, self.num + 1))
            print(f"The sum of all numbers from 1 to {self.num} is: {total_sum}")
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    try:
        user_input = int(input("Enter a number: "))
        sum_calculator = NumberSumCalculator(user_input)
        sum_calculator.calculate_sum()
    except ValueError:
        logging.error("Error: Please enter a valid integer.")
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
```

```python
import logging

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class CharacterSquarePrinter:
    def __init__(self, character, size):
        self.character = character
        self.size = size

    def print_square(self):
        try:
            for _ in range(self.size):
                print(self.character * self.size)
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    try:
        user_char = input("Enter a character: ")[0]  # Take only the first character entered
        user_size = int(input("Enter the number of rows and columns: "))
        
        square_printer = CharacterSquarePrinter(user_char, user_size)
        square_printer.print_square()
    except IndexError:
        logging.error("Error: Please enter a valid character.")
    except ValueError:
        logging.error("Error: Please enter a valid integer for the size.")
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
```

```python
import logging
import math

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class FactorialCalculator:
    def __init__(self, num):
        self.num = num

    def calculate_factorial(self):
        try:
            if self.num < 0:
                raise ValueError("Please enter a non-negative integer.")
            result = math.factorial(self.num)
            print(f"The factorial of {self.num} is: {result}")
        except ValueError as ve:
            logging.error(f"Error: {ve}")
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    try:
        user_input = int(input("Enter a positive integer: "))
        factorial_calculator = FactorialCalculator(user_input)
        factorial_calculator.calculate_factorial()
    except ValueError:
        logging.error("Error: Please enter a valid positive integer.")
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
```

```python
import logging

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class BMICalculator:
    def __init__(self, weight, height):
        self.weight = weight
        self.height = height

    def calculate_bmi(self):
        try:
            if self.weight <= 0 or self.height <= 0:
                raise ValueError("Weight and height must be positive values.")
            
            # Calculate BMI
            bmi = self.weight / (self.height ** 2)
            return bmi
        except ValueError as ve:
            logging.error(f"Error: {ve}")
            return None
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")
            return None

    def determine_bmi_category(self, bmi):
        if bmi is not None:
            if bmi < 18.5:
                return "Underweight"
            elif 18.5 <= bmi < 24.9:
                return "Normal weight"
            elif 25 <= bmi < 29.9:
                return "Overweight"
            else:
                return "Obese"
        else:
            return "BMI calculation failed."

if __name__ == "__main__":
    try:
        user_weight = float(input("Enter your weight in kilograms: "))
        user_height = float(input("Enter your height in meters: "))
        
        bmi_calculator = BMICalculator(user_weight, user_height)
        user_bmi = bmi_calculator.calculate_bmi()

        if user_bmi is not None:
            bmi_category = bmi_calculator.determine_bmi_category(user_bmi)
            print(f"Your BMI is: {user_bmi:.2f}")
            print(f"BMI Category: {bmi_category}")
    except ValueError:
        logging.error("Error: Please enter valid numerical values for weight and height.")
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
```

```python
import logging

logging.basicConfig(level=logging.ERROR)  # Set the desired logging level

class SimpleInterestCalculator:
    def __init__(self, principal, rate, time):
        self.principal = principal
        self.rate = rate
        self.time = time

    def calculate_simple_interest(self):
        try:
            if self.principal <= 0 or self.rate < 0 or self.time <= 0:
                raise ValueError("Principal, rate, and time must be positive values.")
            
            # Calculate Simple Interest
            simple_interest = (self.principal * self.rate * self.time) / 100
            return simple_interest
        except ValueError as ve:
            logging.error(f"Error: {ve}")
            return None
        except Exception as e:
            logging.error(f"An unexpected error occurred: {e}")
            return None

if __name__ == "__main__":
    try:
        user_principal = float(input("Enter the principal amount: "))
        user_rate = float(input("Enter the rate of interest (in percentage): "))
        user_time = float(input("Enter the time in years: "))
        
        interest_calculator = SimpleInterestCalculator(user_principal, user_rate, user_time)
        simple_interest = interest_calculator.calculate_simple_interest()

        if simple_interest is not None:
            print(f"The simple interest is: {simple_interest:.2f}")
    except ValueError:
        logging.error("Error: Please enter valid numerical values for principal, rate, and time.")
    except KeyboardInterrupt:
        logging.error("Program terminated by user.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
```
