1. Write a Python program that takes a string as input and returns the reverse of that
string.
2. Make a function that checks if a given string is a palindrome (reads the same
forwards and backwards).
3. Write a function to count the number of vowels (a, e, i, o, u) in a given string.
4. Given two strings, write a program to concatenate them without using the + operator.
5. Given a string, write a program to extract a substring between a specified start and
end index.
6. Make a function that checks if two given strings are anagrams (contain the same
characters in a different order).
7. Find a function that calculates the frequency of each character in a string and returns
the results as a dictionary.
8. Write a program to remove duplicate characters from a string and return the string
without duplicates.
9. Given two strings, write a function that checks if one is a rotation of the other. For
example, waterbottle is a rotation of erbottlewat.
10. Assume a function to convert a given string into a title case, where the first letter of
each word is capitalized

```python
import logging

# Configure logging
logging.basicConfig(filename='error.log', level=logging.ERROR)

# Problem 1: Reverse a string
def reverse_string(input_string):
    try:
        reversed_str = input_string[::-1]
        return reversed_str
    except Exception as e:
        logging.error(f"Error in reverse_string: {str(e)}")
        return None

# Problem 2: Check if a string is a palindrome
def is_palindrome(input_string):
    try:
        reversed_str = reverse_string(input_string)
        return input_string == reversed_str
    except Exception as e:
        logging.error(f"Error in is_palindrome: {str(e)}")
        return None

# Problem 3: Count the number of vowels in a string
def count_vowels(input_string):
    try:
        vowel_count = sum(1 for char in input_string if char.lower() in 'aeiou')
        return vowel_count
    except Exception as e:
        logging.error(f"Error in count_vowels: {str(e)}")
        return None

# Problem 4: Concatenate strings without using +
def concatenate_strings(str1, str2):
    try:
        concatenated_str = str1 + str2
        return concatenated_str
    except Exception as e:
        logging.error(f"Error in concatenate_strings: {str(e)}")
        return None

# Problem 5: Extract substring between start and end index
def extract_substring(input_string, start_index, end_index):
    try:
        substring = input_string[start_index:end_index+1]
        return substring
    except Exception as e:
        logging.error(f"Error in extract_substring: {str(e)}")
        return None

# Problem 6: Check if two strings are anagrams
def are_anagrams(str1, str2):
    try:
        sorted_str1 = ''.join(sorted(str1))
        sorted_str2 = ''.join(sorted(str2))
        return sorted_str1 == sorted_str2
    except Exception as e:
        logging.error(f"Error in are_anagrams: {str(e)}")
        return None

# Problem 7: Calculate character frequency in a string
def character_frequency(input_string):
    try:
        freq_dict = {}
        for char in input_string:
            freq_dict[char] = freq_dict.get(char, 0) + 1
        return freq_dict
    except Exception as e:
        logging.error(f"Error in character_frequency: {str(e)}")
        return None

# Problem 8: Remove duplicate characters from a string
def remove_duplicates(input_string):
    try:
        unique_chars = ''.join(sorted(set(input_string), key=input_string.index))
        return unique_chars
    except Exception as e:
        logging.error(f"Error in remove_duplicates: {str(e)}")
        return None

# Problem 9: Check if one string is a rotation of the other
def is_rotation(str1, str2):
    try:
        if len(str1) != len(str2):
            return False
        return str1 in (str2 + str2)
    except Exception as e:
        logging.error(f"Error in is_rotation: {str(e)}")
        return None

# Problem 10: Convert string to title case
def title_case(input_string):
    try:
        titlecased_str = ' '.join(word.capitalize() for word in input_string.split())
        return titlecased_str
    except Exception as e:
        logging.error(f"Error in title_case: {str(e)}")
        return None

# Example usage
input_str = "Hello, World!"

print("1. Reverse String:", reverse_string(input_str))
print("2. Is Palindrome:", is_palindrome(input_str))
print("3. Count Vowels:", count_vowels(input_str))
print("4. Concatenate Strings:", concatenate_strings("Hello", "World"))
print("5. Extract Substring:", extract_substring(input_str, 0, 4))
print("6. Are Anagrams:", are_anagrams("listen", "silent"))
print("7. Character Frequency:", character_frequency(input_str))
print("8. Remove Duplicates:", remove_duplicates("Programming"))
print("9. Is Rotation:", is_rotation("waterbottle", "erbottlewat"))
print("10. Title Case:", title_case("hello world"))
```
