```python
class StringManipulator:
    def __init__(self):
        logging.basicConfig(filename='string_manipulation.log', level=logging.ERROR)

    @staticmethod
    def reverse_string(input_str):
        try:
            reversed_str = input_str[::-1]
            return reversed_str
        except Exception as e:
            logging.error(f"Error in reverse_string function: {e}")

    @staticmethod
    def is_palindrome(input_str):
        try:
            return input_str == input_str[::-1]
        except Exception as e:
            logging.error(f"Error in is_palindrome function: {e}")

    @staticmethod
    def count_vowels(input_str):
        try:
            vowels = 'aeiou'
            count = sum(1 for char in input_str if char.lower() in vowels)
            return count
        except Exception as e:
            logging.error(f"Error in count_vowels function: {e}")

    @staticmethod
    def concatenate_strings(str1, str2):
        try:
            concatenated_str = ''.join([str1, str2])
            return concatenated_str
        except Exception as e:
            logging.error(f"Error in concatenate_strings function: {e}")

    @staticmethod
    def extract_substring(input_str, start_index, end_index):
        try:
            substring = input_str[start_index:end_index]
            return substring
        except Exception as e:
            logging.error(f"Error in extract_substring function: {e}")

    @staticmethod
    def are_anagrams(str1, str2):
        try:
            return sorted(str1) == sorted(str2)
        except Exception as e:
            logging.error(f"Error in are_anagrams function: {e}")

    @staticmethod
    def calculate_char_frequency(input_str):
        try:
            char_frequency = {}
            for char in input_str:
                char_frequency[char] = char_frequency.get(char, 0) + 1
            return char_frequency
        except Exception as e:
            logging.error(f"Error in calculate_char_frequency function: {e}")

    @staticmethod
    def remove_duplicates(input_str):
        try:
            unique_chars = ''.join(sorted(set(input_str), key=input_str.index))
            return unique_chars
        except Exception as e:
            logging.error(f"Error in remove_duplicates function: {e}")

    @staticmethod
    def are_rotations(str1, str2):
        try:
            return len(str1) == len(str2) and str2 in str1 + str1
        except Exception as e:
            logging.error(f"Error in are_rotations function: {e}")

    @staticmethod
    def title_case(input_str):
        try:
            title_case_str = ' '.join(word.capitalize() for word in input_str.split())
            return title_case_str
        except Exception as e:
            logging.error(f"Error in title_case function: {e}")


# Example usage:
str1_anagram = "listen"
str2_anagram = "silent"
print(f"Are Anagrams: {StringManipulator.are_anagrams(str1_anagram, str2_anagram)}")

str_with_duplicates = "programming"
print(f"Character Frequency: {StringManipulator.calculate_char_frequency(str_with_duplicates)}")

str_with_duplicates_removed = StringManipulator.remove_duplicates(str_with_duplicates)
print(f"String without Duplicates: {str_with_duplicates_removed}")

rotation_str1 = "waterbottle"
rotation_str2 = "erbottlewat"
print(f"Are Rotations: {StringManipulator.are_rotations(rotation_str1, rotation_str2)}")

title_case_str = "this is a title"
print(f"Title Case: {StringManipulator.title_case(title_case_str)}")
```
