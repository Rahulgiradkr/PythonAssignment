```python
from datetime import datetime, timedelta

def convert_to_minutes(login_time, logout_time):
    # Convert time strings to datetime objects
    login_datetime = datetime.strptime(login_time, "%H:%M:%S")
    logout_datetime = datetime.strptime(logout_time, "%H:%M:%S")

    # Calculate duration
    duration = logout_datetime - login_datetime

    # Convert duration to minutes
    duration_minutes = duration.total_seconds() / 60

    return duration_minutes

# Example usage
login_time = "15:35:29"
logout_time = "17:39:39"
duration_minutes = convert_to_minutes(login_time, logout_time)

print(f"Duration in minutes: {duration_minutes:.2f} minutes")
```
